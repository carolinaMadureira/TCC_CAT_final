```python
import numpy as np
import pandas as pd
from datetime import timedelta
import math

from dateutil.relativedelta import *
from datetime import date
from datetime import datetime

def formatarStr (DataFrame, columns):
    for col in columns:
        DataFrame[col] = DataFrame[col].str.strip()
        
def recortarCNAE(cnaeStr):
    strSeparada = cnaeStr.rsplit(":")
    if len(strSeparada) > 1:
        return strSeparada[0]
    else:
        return cnaeStr

def calcularSomaCNAE(cnaeStr):
    try: 
        df_soma = df_result.query("cnae == \""+cnaeStr+"\"" )
        total= df_soma['qtde'].sum()    
        return total
    except (RuntimeError, TypeError, NameError, ValueError):
        return np.nan
```

```python
#ISO-8859-1
nomeArquivoOrigemCNAE = 'datasets/origem/MPS/ACT10.csv'
nomeArquivoResultadoCNAETotal = 'datasets/result/dados_CNAE_MPS_t.csv'
nomeArquivoResultadoCNAE = 'datasets/result/dados_CNAE_MPS.csv'

columns_types = {'ano': str, 'cnae': str, 'motivo"': str, 'qtde': str, 'situacao': str, 'grupo_cnae': str, 
                 'grupo_cnae_letra': str}

nomes  = ['ano', 'cnae', 'motivo', 'qtde', 'situacao', 'grupo_cnae', 'grupo_cnae_letra']

dados_originais = pd.read_csv(nomeArquivoOrigemCNAE, sep = ',', header=0, names=nomes, dtype=columns_types, encoding='ISO-8859-1')

df_original = pd.DataFrame(data = dados_originais, columns=nomes)
df_original.shape[0] #34168
```

```python
df_original.head()
```

```python
df_original.drop(['motivo'], axis=1, inplace = True)
formatarStr(df_original, ['ano', 'cnae', 'qtde', 'situacao', 'grupo_cnae', 'grupo_cnae_letra'])

selecaoRegistrados = df_original['situacao'] == 'Com Cat Registrada'
df_original = df_original[selecaoRegistrados]
df_original.shape[0] #24444
```

```python
df_original.drop(['situacao'], axis=1, inplace = True)
df_original.drop(['grupo_cnae'], axis=1, inplace = True)
df_original.drop(['grupo_cnae_letra'], axis=1, inplace = True)

df_original.dropna(subset = ['qtde'], inplace = True)
df_original.dropna(subset = ['ano'], inplace = True)

df_original['ano'] = df_original['ano'].apply(pd.to_numeric)
df_original['qtde'] = df_original['qtde'].apply(pd.to_numeric)

selecao2016 = df_original['ano'] >= 2016
df_original = df_original[selecao2016]

#excluir os registros cujo cbo foi preenchido com 'Ignorado'
selecaoIgnorado = df_original['cnae'] != '9999:Ignorado'
df_original = df_original[selecaoIgnorado]
df_original.shape[0] #5511
```

```python
df_original.info()
```

```python
df_original.index = range(df_original.shape[0]) #0:5511
df_original['cnae'] = df_original['cnae'].apply(lambda x: recortarCNAE(x))

df_original.head(20)
```

```python
df_result = pd.DataFrame(df_original.groupby(['cnae', 'ano'], as_index=False)['qtde'].sum())
df_result.head(30)
```

```python
df_result.shape[0] #1993
```

```python
df_result.to_csv(nomeArquivoResultadoCNAE, sep = ';', encoding='UTF-8')
```

```python
df_resultGroup = pd.DataFrame({'cnae': df_result['cnae']})
df_resultGroup.drop_duplicates(subset ="cnae", keep = 'first', inplace = True)
df_resultGroup['total'] = df_resultGroup['cnae'].apply(lambda x: calcularSomaCNAE(x))
df_resultGroup = df_resultGroup.sort_values(['total'], ascending = False)

df_resultGroup.index = range(df_resultGroup.shape[0]) #0:47
df_resultGroup.head(10)
```
